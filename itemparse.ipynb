```python
datas = {}
checkdup = {
    'statAttackPowerRanged': 'statAttackPower',
}

def get_spell_effects(spid):
    out = {}
    sp = spells[spid]
    for i in range(1, 5):
        k = 'effectapplyauraname' + str(i)
        if k not in sp:
            continue
        hasaura = True
        v = 'effectbasepoints' + str(i)
        ea = speas.get(sp[k], str(sp[k]))
        if ea == 'haste':
#             emv[sp['effectmiscvalue' + str(i)]] = sp['description1']
            ea = effmisc.get(sp['effectmiscvalue' + str(i)]) or 'unk'
            if ea == 'unk':
                print(data['entry'], sp['effectmiscvalue' + str(i)], sp['description1'])
        seenspeas.add(sp[k])
        vv = sp.get(v, 0) + sp.get('effectdiesides' + str(i), 0)
        kk = 'stat'+ ea[0].upper() + ea[1:]
        out[kk]  = out.get(kk, 0) + vv

    for k, v in list(out.items()):
        if k in checkdup:
            if v == out.get(checkdup[k]):
                del out[k]

    return out

def apply_prop(o, k, v):
    if v:
        o[k] = v    

def pull_prop(o, k):
    t = None
    if k in o:
        t = o[k]
        del o[k]
    return t    

renames = {
    'entry': 'id',
    'itemlevel': 'ilvl',
    'dmg_min1': 'dmgMin',
    'dmg_max1': 'dmgMax',
    'delay': 'speed',
    'requiredlevel': 'rlvl',
}

for line in open('item_template.json'):
    data = json.loads(line)
    if data.get('itemlevel', 0) < 60 or (data.get('requiredlevel', 0) < 60 and 'requiredlevel' in data):
        continue
    clv = '%s|%s' % (data.get('class', 0), data.get('subclass', 0))
    if data.get('class') not in types:
        continue
    data['type'] = classes.get(clv, clv)
    data['slot'] = slots.get(data.get('inventorytype'), 0)

    datas[data['entry']] = data

    if 'spells' not in data:
        data['spells'] = []

    for i in range(1, 6):
        k = 'socketcolor_' + str(i)
        if k not in data:
            continue
        if 'sockets' not in data:
            data['sockets'] = []
        skt = pull_prop(data, k)
        data['sockets'].append(sockets.get(skt, skt))
    
    if 'socketbonus' in data:
        e = enchants[data['socketbonus']]
#         if e['spell'] in gemprops:
#             e = gemprops[e['spell']]
#         spid = e['spell']
#         data['socketbonus'] = get_spell_effects(spid) or spid
        data['socketbonus'] = sockbon[e['name']]
    
    if 'gemproperties' in data:
        e = gemprops[data['gemproperties']]['enchant']
        e = enchants[e]
#         print(data['entry'], data['name'], data['gemproperties'], e)
        data['gemproperties'] = e['name']
        
    
    for i in range(1, 11):
        k = 'stat_type' + str(i)
        v = 'stat_value' + str(i)
        if k in data or v in data:
            kk = data.get(k, 0)
            vv = data.get(v, 0)
            if k in data:
                del data[k]
            if v in data:
                del data[v]
            if kk not in stats:
                print('unkstat', data['entry'], kk, vv)
            else:
                kk = to_camel_case('stat_'+ stats[kk])
                data[kk]  = data.get(kk, 0) + vv

    for i in range(1, 5):
        k = 'spellid_' + str(i)
        if k not in data:
            continue
        spid = data[k]

        t = 'spelltrigger_' + str(i)
        tr = pull_prop(data, t)
        ppm = 'spellppmrate_' + str(i)
        cd = 'spellcooldown_' + str(i)
        active = 'spellcategorycooldown_' + str(i)
        if t in data and tr != 1 or ppm in data or cd in data:
            se = get_spell_effects(spid)
            apply_prop(se, 'trigger', tr)
            if 'trigger' in se:
                se['trigger'] = trigs[se['trigger']]
            apply_prop(se, 'cd', (pull_prop(data, cd) or 0) // 1000)
            apply_prop(se, 'ppm', pull_prop(data, ppm))
            if 'cd' in se and 'ppm' in se:
                del se['ppm']
            apply_prop(se, 'duration', (pull_prop(data, active) or 0) // 1000)
            if 'duration' in se and se.get('duration') == se.get('cd'):
                del se['duration']
            data['spells'].append(se)
            del data[k]
            continue
        del data[k]
#         hasaura = False
        se = get_spell_effects(spid)
        for k, v in se.items():
            if data.get(k) == (v+1):
                continue
            data[k] = data.get(k, 0) + v +1
#             hasaura = True
#         if not hasaura:
#             data['spells'].append({
#                 'description': sp['description1']
#             })

    if not data['spells']:
        del data['spells']
                
                
    for k, v in list(data.items()):
        if k in checkdup:
            if v == data.get(checkdup[k]):
                del data[k]

    if data['entry'] in loot:
        ph = loot[data['entry']].get('phase')
        if ph:
            data['phase'] = int(ph)
                
    for k, v in renames.items():
        if k in data:
            data[v] = data[k]
            del data[k]
    
                
#     print(data)
#     break
len(datas.keys())
```
